```python
from jinja2 import Template
```

# People places

```python
people_places = [
    ["Vidu", "HBH"],
    ["Rahul", "Bhadohi"],
    ["Harikesh", "Bhadohi"],
    ["Muskan", "Raipur"],
    ["Vaish", "Indore"]
]
```

```python
template_string = '''
<div class='container'>
  {% for fruit in ['apple','orange','banana'] %}
    {{fruit}}
  {% endfor %}
</div>
'''

template = Template(template_string.strip())
markup = template.render()
print(markup)
```

```python
template_string = '''
<div class='container'>
  {% for name, place in my_data %}
    <p>I am {{name}}, I live in {{place}}</p>
  {% endfor %}
</div>
'''

template = Template(template_string.strip())
markup = template.render(my_data=people_places)
print(markup)
```

```python
template_string = '''
<div class='container'>
  {% for name, place in my_data %}
    <p>
      I am {{name}}, I live in {{place}}
    </p>
  {% endfor %}
</div>
'''

template = Template(template_string.strip())
markup = template.render(my_data=people_places)
print(markup)
```

# Todos

```python
todos = [
    {"text": "Learn good styling", "status": "done"},
    {"text": "Learn varna of raga kamboji", "status": "pending"},
    {"text": "Learn", "status": "pending"},
    {"text": "Fly in an airplane", "status": "pending"},
    {"text": "Fly in a hot air balloon", "status": "pending"},
    {"text": "Swim in the ocean", "status": "done"},
    {"text": "Sky jump", "status": "pending"},
    {"text": "Travel to Norway", "status": "pending"}
]
```

```python
template_string = '''
<div class='container'>
  {% for todo in todos %}
    <p class='todo {{todo.status}}'>
      {% if todo.status == 'done' %}
      ✅ {{todo.text}}
      {% elif todo.status == 'pending' %}
      ⌛ {{todo.text}}
      {% endif %}
    </p>
  {% endfor %}
</div>
'''

template = Template(template_string)
markup = template.render(todos=todos)
print(markup)
```

```python
template_string = '''
<div class= 'container'>
  {% for todo in todos %}
  <p class='todo {{todo.status}}'>
    {% if todo.status == 'done' %}
    ✅ {{todo.text}}
    {% elif todo.status == 'pending' %}
    ⌛ {{todo.text}}
    {% endif %}
  </p>
  {% endfor %}
</div>
'''

template = Template(template_string)
markup = template.render(todos=todos)
print(markup)
```
